```python
# import  libraries 
import pandas as pd
import numpy as np
```

```python
# load dataset
ds=pd.read_csv("train.csv",parse_dates=['Order Date','Ship Date'],dayfirst=True)
```

```python
# analysis
ds.info()
```

```python
ds.describe()
```

```python
#Count null values
ds["Postal Code"].isnull().value_counts()
```

```python
#1.Remove null values
ds.dropna(subset=["Postal Code"],inplace=True)
```

```python
ds.info()
```

```python
ds.head(20)
```

```python
#Change the data type
ds['Ship Mode'] = ds['Ship Mode'].astype(str)
ds['Customer ID'] = ds['Customer ID'].astype(str)
ds['Segment'] = ds['Segment'].astype(str)
ds['Country'] = ds['Country'].astype(str)
ds['City'] = ds['City'].astype(str)
ds['State'] = ds['State'].astype(str)
ds['Region'] = ds['Region'].astype(str)
ds['Product ID'] = ds['Product ID'].astype(str)
ds['Category'] = ds['Category'].astype(str)
ds['Sub-Category'] = ds['Sub-Category'].astype(str)
ds['Product Name'] = ds['Product Name'].astype(str)
```

```python
ds.info()
```

```python
#Count duplicated values
ds.duplicated().value_counts()
```

```python
#Change the formate of column name
ds.columns=ds.columns.str.lower()
ds.columns=ds.columns.str.replace(" ","_")
ds.columns=ds.columns.str.replace("-","_")
ds=ds.rename(columns={'row_id_(usd)':'row_id'})
```

```python
ds.to_csv("Cleaned_data.csv")
```
