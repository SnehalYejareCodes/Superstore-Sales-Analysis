```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
ds=pd.read_csv("Cleaned_data.csv")
```

```python
ds
```

```python
ds = ds.drop('Unnamed: 0', axis=1)
```

```python
username = 'root'
password = 'password'
host = 'localhost'  # or your host IP
port = '3306'       # default MySQL port
database = 'superstore_db'
```

```python
# STEP 3: Create connection engine
engine = create_engine(f"mysql+mysqlconnector://{username}:{password}@{host}:{port}/{database}")
```

```python
# STEP 4: Write DataFrame to MySQL table
try:
    ds.to_sql(name='superstore_sales', con=engine, if_exists='replace', index=False)
    print("✅ Data transferred successfully to MySQL!")
except Exception as e:
    print("❌ Error transferring data to MySQL:", e)
```
